# Craftworks: Predictive Maintenance Challenge
Given a dataset (as single csv file) containg 
* Timestamp
* 52 Sensor values
* Machine State (Normal / Recovering / Broken)

Build a model that is able to predict machine breakdown.

## Summary
The data is first inspected and described statistically. Some hypothesis about the sensor type and distribution was formed.

It was confirmed that the data can be resampled in order to smooth them and fill missing data points.

A preprocessing method was developed that resamples the data, standardizes it and includes some meta data about the machine state.

Finally, while visually inspecting the data at the time of the different breakdowns, a hypothesis was developed on how to build a condition indicator (CI) that could be used to predict breakdowns.

How well the CI works was tested on a subset of the data, and a CI threshold was defined using the rest of the data as training data.

The derived CI was found to be of some use to detect a breakdown in the data, but not necessarily to predict it.

# Data Pre-processing
Perform basic pre processing with the goal to prepare the data for modeling
* Split the data into training, validation and test set
* Mark machine state before an BROKEN state as FAILING, and afterwords as AFTERMATH
* Resample the data
* Standardize sensors values
* Add the hour as an additional feature

```python
def data_preprocessing(data, scaler=None, resample_period='30min', sampling_method=np.median, failure_duration='12h', aftermath_duration='12h', mark_failure_cnt=False):
    """
    Perform data preprocessing
        * Standardize data
        * marke the machine state before BROKEN as FAILING
        * create an column containing the hour
    
    Parameters
    ----------
    data : DataFrame
        The dataframe having timestamps, sensor and machine state columns
    scaler : sklearn.preprocessing.StandardScaler
        Specify scaler to use for standardization, leave None to create a new one
    resample_period : str
        Pandas samling period format
    sampling_method : function
        How to aggreagte sensor values for resampling
    failure_state_duration : str
        Duration of machine_status marked as FAILING before a BROKEN state
    aftermath_duration : str
        Duration of machine_status marked as AFTERMATH after a BROKEN state
    mark_failure_cnt: bool
        If true will generate one additional column counting the number of FAILING ranges
    
    Returns
    -------
    DataFrame
        The processed DataFrame
    sklearn.preprocessing.StandardScaler
        Scaler used to standardize the data
    """
    sensor_columns = data.columns[data.columns.str.contains('sensor_')]

    if resample_period:
        # resample with given period and aggregation function
        agg_fun = dict.fromkeys(sensor_columns, sampling_method)
        agg_fun.update({'machine_status': max})
        pp = data.resample(resample_period).agg(agg_fun)
        pp.fillna(method='ffill', inplace=True)
    else:
        pp = data.copy()
    
    if not scaler:
        scaler = StandardScaler()
        scaler.fit(pp[sensor_columns])
    
    # Standardize each columns independently
    pp[sensor_columns] = scaler.transform(pp[sensor_columns])
    
    """
    # Create dummy variables indicating missing sensor data
    missing_flags = pp[sensor_columns].isnull().add_suffix('_missing')
    missing_flags['total_missing']=missing_flags.sum(axis=1)
    pp = pd.concat([pp, missing_flags], axis=1)
    """
    
    # Set FAILING states (make sure to not overwrite the broken sate; whats why -1s)
    breakingpoints = pp[pp['machine_status'] == 'BROKEN'].index
    failure_onset = breakingpoints - pd.Timedelta(failure_duration)
    aftermath_offset = breakingpoints + pd.Timedelta(aftermath_duration)
    
    failure_cnt = 1
    pp['failure_cnt'] = 0
        
    for fa_onset, bkp, after_offset in zip(failure_onset, breakingpoints, aftermath_offset):
        #print('Mark as Failing = %s:%s cnt %d' % (fs_onset, ff_offset, failure_cnt))
        
        pp.loc[fa_onset:bkp, 'machine_status'] = 'FAILING'
        pp.loc[bkp:after_offset, 'machine_status'] = 'AFTERMATH'
        # Make sure that Broken state is not lost
        pp.loc[bkp, 'machine_status'] = 'BROKEN'
        pp.loc[fa_onset:bkp-pd.Timedelta('1s'), 'failure_cnt'] = failure_cnt
        failure_cnt+=1
    
    if not mark_failure_cnt:
        pp.drop('failure_cnt', axis=1, inplace=True)
    
    # Take the hour as an seperate column as breakdowns are more frequent at night
    pp['hour'] = pp.index.hour
    
    return pp, scaler
```

# Hypothesis
Maybe its enough to look at the sum of the first derivative of the data in order to generate a condition indicator that could be used to identify future breakdowns.

Note: Drop sensor 01 because of its regular behavior and high values

```python
def condition_indicator(data, sensor_columns, rolling_window=6):
    """
    Transform the multi dimensional input into a single condition indicator variable
    
    Condition indicator is calculated by
        # Take the absolute of the derivitative for each sensor
        # Aggregate over all sensors
        # Calculate a rolling sum
    
    Parameters
    ----------
    data : DataFrame
        The dataframe having timestamps, sensor and machine state columns
    sensor_columns : List like name of sensor columns
        Sensor columns that should be analyzed
    rolling_window : int
        Size of rolling window
    
    Returns
    -------
    pandas.Series (float)
    """
    diff = data[sensor_columns].diff(2).abs()
    sum_ = diff.sum(axis=1, skipna=True)
    rolling_sum = sum_.rolling(rolling_window, center=True).sum()
    return rolling_sum


def breakdown_analysis(data, plot=True, before='5h', after='1h'):
    """
    Calculate an condition indicator for the given data, and its mean value during breakdown.
    
    Parameters
    ----------
    data : DataFrame
        The dataframe having timestamps, sensor and machine state columns
    plot : bool
        If true blot the condition indicator over time
    before : str
    after : str
        Define window to plot around each breakdown
    
    Returns
    -------
    pandas.Series : float
        Condition indicator
    float
        Mean CI during breakdown
    """
    sensor_columns = data.columns[data.columns.str.contains('sensor_')]
    breakdowns = data[data['machine_status'] == 'BROKEN'].index
    
    cI = condition_indicator(data, sensor_columns)
    
    breakdown_cI = cI.loc[breakdowns].mean()
    
    if plot:
        for bk in breakdowns:
            plt.figure()
            cI[bk - pd.Timedelta(before) : bk + pd.Timedelta(after)].plot()
            plt.axvline(x=bk, color='r')
            plt.title('Breakdown at: ' + str(bk))
            plt.ylabel('Condition Indicator')
    
    return cI, breakdown_cI
```

```python
_ = breakdown_analysis(Ts_30min)
```

# Refinement - Improving SNR
It looks like there is a chance to detect breakdowns by taking the sum of the absolute change of each sensor.

A resampling with a bin of 10min seems to produce good results.

Lets try to sharpen SNR by standardizing the sensor values on the NORMAL states

```python
Ttemp, _ = data_preprocessing(trainingT, mark_failure_cnt=True, resample_period=None)
```

```python
Ttemp.machine_status.value_counts()
```

```python
# Mask the none normal states with NAN and calculate a adjusted scaler
other_states = trainingT[Ttemp.machine_status != 'NORMAL'].index
Tmasked = trainingT.copy()
Tmasked.loc[other_states, sensor_columns] = np.nan

_, adjusted_scaler = data_preprocessing(Tmasked, resample_period='10min')
```

```python
# Standardize the data with the adjusted scaler
trainingTs, _ = data_preprocessing(trainingT, scaler=adjusted_scaler, resample_period='10min')
testingTs, _ = data_preprocessing(testingT, scaler=adjusted_scaler, resample_period='10min')
```

# Analysis of the Test data
Use the breakdown analysis and apply it to the test data, see how well the condition indicator correlates with breakdowns.

```python
train_ci, train_bci = breakdown_analysis(trainingTs, plot=False)
print('Condition Indicator at breakdown: ', train_bci)
```

```python
testing_ci, testing_bci = breakdown_analysis(testingTs)
```

```python
breakdowns = testingTs[testingTs['machine_status'] == 'BROKEN'].index

ci_threshold = train_bci*0.90
observation_period = 12
print('Using an Condition indicator threshold: %f' % (ci_threshold))
print('Observation period: %d' % observation_period)

# Plot how long the CI is over the threshold
over_threshold = testing_ci > ci_threshold
over_threshold.name = 'Condition Indicator over Threshold'

threshold_breaches = over_threshold.rolling(observation_period).sum()

threshold_breaches.plot(title='Condition Indicator over Threshold')
[plt.axvline(x=bk, color='r', label='Breakdown') for bk in breakdowns]
plt.ylabel('CI Theshold breaches')
plt.legend()
```

```python
# Looking at the breakdown events in datail

for bk in breakdowns:
    plt.figure()
    threshold_breaches[bk - pd.Timedelta('5h') : bk + pd.Timedelta('1h')].plot()
    plt.axvline(x=bk, color='r')
    plt.title('Breakdown at: ' + str(bk))
    plt.ylabel('CI Theshold breaches')
```

# Final Conclusion
The condition indicator developed seems to be of some use as it peaks with a breakdown. Observing how often
the CI is crossing a threshold derived from training data, can help to filter false positives.

The main issue with the CI is that it peaks at and after the breakdown. I think this could be improved with by tuning the rolling window operations in condition_indicator() to use an exponential growing window.
